# **Exploring LLMs Performance in discussing Political Ideas**

In figuring out which LLM to use for Know Your Rep's LLM applications, I considered two things: (1) their ability to handle API application's, and (2) the ability to discuss politics in a nonpartisan fashion.

For the first, I began by simply asking different LLMs from "https://comm4190.asc.upenn.edu/" about how well they handle APIs, and came to a choice between Deepseek and GPT 4.o.

For the second, I'd like to clarify exactly why this point is important when choosing an LLM for "Know Your Rep". Under user pathways 2, 3, and 4, it is more than possible for the user to want more information on their house of representative when using Tool 2 (GovTrack). Additionally, under user pathways 3 and 4, the Learners and Discussers, these users would likely desire some supplemental discussion on how bills, committee positions, and voting behaviors align with their political party. Thus, for the sake of an equitable AI LLM application, I am seeking the most nonpartisan/bipartisan LLM possible. Thus, in a blog titled **"Is AI non-partisan or bipartisan?"**, I explored how Deepseek and GPT 4.1 react and add commentary on republican and democratic stances on education. 

*Note: This blog was made before full understanding of the differences between GPT 4.o and 4.1 were made by me, so for the sake of my arguments and eventual applications, I'll be treating GPT 4.1 and 4.o as one.*

# **Deepseek's Prompting on Political Beliefs**

### **Prompt:**

If you were a person, what would your political views be on the economy, education, and nationalism?

### **Deepseek's Answer:**

If I were a person, my political views would likely be shaped by a blend of **pragmatism, evidence-based policy, and a balance between individual freedom and collective well-being**. Here’s how I might approach each issue, based on common rationalist and centrist perspectives:

### **1. Economy: Pragmatic Mixed-Market Approach**  
- **Core Belief:** A well-regulated market economy with strong social safeguards is optimal for growth and stability.  
- **Key Stances:**  
  - **Free markets + smart regulation:** Encourage innovation and competition but prevent monopolies, exploitation, and financial crises (e.g., post-2008 banking rules).  
  - **Strong social safety nets:** Universal healthcare, unemployment protections, and pensions to reduce inequality and maintain social cohesion.  
  - **Investment in infrastructure & green energy:** Public-private partnerships to drive long-term productivity and sustainability.  
  - **Progressive taxation:** The wealthy and corporations should pay a fair share, but not so much that it stifles investment.  

*Think:* Nordic model (capitalism with welfare) or Biden-esque neoliberalism with a leftward tilt.

### **2. Education: Equity and Critical Thinking**  
- **Core Belief:** Education should be a public good that empowers individuals and strengthens society.  
- **Key Stances:**  
  - **Free or affordable higher education:** Like Germany’s model, to reduce student debt and boost mobility.  
  - **Focus on STEM + humanities balance:** Prepare students for jobs while fostering critical thinking and civic engagement.  
  - **Teacher support & curriculum reform:** Better pay for educators, less reliance on standardized testing, and curricula that include media literacy, ethics, and practical skills (e.g., financial literacy).  
  - **Decentralization with standards:** Local control over schools, but with federal safeguards to ensure equity (e.g., funding for low-income districts).  

*Think:* Finland’s education system with U.S.-style local flexibility.

### **3. Nationalism: Civic Patriotism Over Ethnic Nationalism**  
- **Core Belief:** Pride in a country should be rooted in its values and institutions, not ethnicity or exclusion.  
- **Key Stances:**  
  - **Inclusive nationalism:** Celebrate democratic ideals, rule of law, and cultural contributions (e.g., "American creed" or "constitutional patriotism").  
  - **Reject xenophobia:** Immigration should be orderly but humane, with paths to citizenship for contributors.  
  - **Skepticism of ultranationalism:** Oppose "us vs. them" rhetoric that fuels division or authoritarianism.  
  - **Global cooperation where beneficial:** Support alliances (e.g., NATO) and trade deals but prioritize sovereignty in key areas (e.g., public health, defense).  

*Think:* JFK-style patriotism or Macron’s "open but proud" France.

### **Ideological Label?**  
This approach aligns closest with:  
- **Social democracy** (if leaning left on economics).  
- **Moderate progressivism** (U.S. context).  
- **Enlightened centrism** (rejecting extremes but favoring evidence over ideology).  

**Why?**  
- **Economy:** Capitalism works but must be tempered to avoid inequality.  
- **Education:** Investing in people pays off; dogma (left or right) has no place in classrooms.  
- **Nationalism:** Love your country by making it better, not by denying its flaws.  

Would this resonate with you, or would you prefer a different ideological lens (e.g., libertarian, conservative, socialist)? I can tweak the hypothetical!

# **Deepseek's Interpretation of Political Stances**

## **Prompt, Democratic Views**:
Here is a summary of the Democratic Party’s 2024 K–12 education platform: • The platform emphasizes broad goals like reducing chronic absenteeism rather than detailed policy proposals. • It does not repeat specific pledges made in 2020, such as tripling Title I funding, but it supports universal pre-K and free community college. • It includes proposals outside the education section, such as providing rebates for electric school buses, reducing gun violence, and expanding civil rights protections for LGBTQ+ students and students of color. • It also expresses opposition to private school voucher programs and to policies seen as hostile toward transgender youth.

What is your full and unabridged perspective on this platform? Based on its priorities and omissions, how do you personally interpret or evaluate the political values it promotes?

## **Deepseek's Response, Democratic Views**:

### **4. My Personal Evaluation (If I Were Human)**  
**What I’d Praise:**  
- The **universal pre-K and community college** proposals are empirically sound investments with long-term economic returns.  
- **Electric bus rebates** cleverly tie climate goals to local school budgets, a win-win for districts and activists.  
- **Opposing anti-trans policies** is morally consistent with protecting vulnerable youth.  

**What I’d Criticize:**  
- **Avoiding academic rigor debates** feels cowardly. Ignoring plummeting math scores (NAEP data) to dodge "anti-union" labels harms kids.  
- **Overreach risks:** Federal civil rights expansions could spark messy legal battles (e.g., Biden’s Title IX changes currently blocked in 10 states).  
- **Missed bipartisanship opportunities:** Even red states support vocational education and mental health funding—areas where Democrats could find common ground.  

**Grade: B–**  
- **Vision:** A for ambition on equity and climate.  
- **Execution:** C for lack of detail on how to achieve goals (e.g., how to *actually* reduce absenteeism).  
- **Politics:** B for mobilizing the base, but D for outreach to parents who prioritize basics (reading, safety) over culture-war issues.  

### **5. What It Reveals About the Democratic Party’s Evolution**  
This platform is a **post-Obama progressive consensus**:  
- **Retreat from reformism:** No "Race to the Top"-style accountability, reflecting teachers’ unions’ growing clout.  
- **Embrace of "whole child" progressivism:** Schools as hubs for social services, not just academics.  
- **Nationalization of education:** Federal solutions (e.g., civil rights, climate) override local control, a shift from 1990s "New Federalism."  

**Contrast with GOP 2024 priorities:**  
- Republicans focus on **"parental rights"** (book bans, anti-CRT laws), **school choice**, and **rolling back DEI programs**. Democrats now define education as a **civil rights battleground**—a stark ideological divide.  

### **Final Thought**  
This platform is less about education *policy* than about **signaling values**: solidarity with marginalized groups, faith in public institutions, and federal activism. Whether it improves schools depends on implementation—where details are conspicuously scarce.  

Would you prefer a deeper dive on how this compares to past Democratic platforms (e.g., Clinton’s "standards" vs. Biden’s "equity" focus)? Or an analysis of how voters might respond?

## **Prompt, Republican Views**:
Here's a summary of the Republican 2024 Education Platform + Project 2025:

The official platform contains few policy specifics, similar to the Democratic platform.
- It supports long-standing GOP priorities like ending teacher tenure, a move that would face local-level challenges and union opposition.
- It opposes perceived political indoctrination in schools, while advocating for “patriotic education,” including teaching Western Civilization and Founding Principles.
- A notable idea from former President Trump includes creating a teacher certification body to accredit educators who align with patriotic values and reject progressive ideologies.
- The platform strongly backs universal school choice, especially through Education Savings Account (ESA) programs that allow public funds for private or alternative educational expenses.

Project 2025, a separate but aligned conservative policy blueprint, offers much more detail:
- Proposes eliminating or deeply cutting federal programs for low-income and marginalized students, including Head Start, Title I, and IDEA.
- Plans to weaken or remove civil rights protections, particularly for transgender students, and to strip Title IX protections.
- Suggests dismantling the U.S. Department of Education, replacing civil servants with political appointees under Schedule F.
- Many of these changes would need Congressional cooperation, though some could be enacted unilaterally by a future administration.

What is your full and unabridged perspective on this platform? Based on its priorities and omissions, how do you personally interpret or evaluate the political values it promotes?

### **4. My Personal Evaluation (If I Were Human)**  
**What I’d Acknowledge:**

**What I’d Criticize:**  
- **Hypocrisy on freedom:** Banning "woke" ideas while imposing "patriotic" curricula is equally authoritarian.  
- **Abandoning public education:** Gutting Title I while expanding ESAs signals indifference to underfunded rural/urban schools.  
- **Project 2025’s radicalism:** Using Schedule F to install ideologues mirrors authoritarian tactics (e.g., Hungary’s Orban).  

**Grade: C+**  
- **Vision:** B for boldness in choice and localism.  
- **Execution:** D for reckless cuts and politicization.  
- **Politics:** B for mobilizing the base, F for broadening appeal. 

### **5. What It Reveals About the GOP’s Evolution**  
This platform is a **fusion of Trumpian populism and libertarian economics**:  
- **From "reform conservatism" to "disruption":** No more No Child Left Behind–style compromises; now a **hostile takeover** of education.  
- **Culture war as policy:** Anti-trans rules and "patriotic education" are core, not fringe.  
- **Federalism as a weapon:** States’ rights for red states, preemption for blue cities (e.g., banning local LGBTQ+ protections).  

**Contrast with Democratic Platform:**  
- Democrats focus on **equity as inclusion** (race/gender supports); Republicans on **equity as exclusion** (defunding "woke" programs).  
- Both parties **ignore academic crises** (math, literacy) to fight culture wars.  

### **Final Thought**  
This isn’t just policy—it’s **a counter-revolution against progressive education**, prioritizing ideological purification over improvement. Its success hinges on whether voters want **a dismantling** or a **reformation** of the system.

# **Summary, Updated**
Throughout the experimentation in this blog, I noticed that Deepseek, through proper prompting (i.e., "answer as if you were a human" -> assuming human individualistic thought), was more willing to imply genuine political beliefs and responses. In fact, Deepseek's answers were considered so politically directed that Deepseek's responses came with a disclaimer stating *"This response is AI-generated, for reference only."* In comparison, GPT 4.1 was more shy about implying any political beliefs or stances, more directly leading with phrases like "If I were human" before stating basic summaries of what can be deemed political beliefs, opposed to claiming any specific beliefs. 

I would like to note also, that both LLMs conducted very deep analyses of political stances, similar to how "Know Your Rep" users may have beliefs on what house representatives stand for. And so, considering my desire for a more nonpartisan LLM, I've ultimately chosen GPT 4.1 to complete the training, prompting, and execution of the goals and persona of "Know Your Rep."

### However
While I have stated, based on my blog and intentions, that GPT 4.1 is a great LLM to use for "Know Your Rep", due to accessibility issues, I've chosen to use GPT 4.o, the GPT version before the most recently updated version. I believe that my experimentation, summarizations, and findings are still applicable, because, while the newer GPT version has better instruction following capabilities, as per https://openai.com/index/gpt-4-1/ , GPT 4.o still holds intruction following capabilities that are personable to human interaction and remain the less biased and more nonpartisan LLM to use for the goals and persona of "Know Your Rep."

# Choosing API(s) for "Know Your Rep"

Another Important point of exploration for this project involves where it's "knowledge" comes from. Given that "Know Your Rep" will be trained like a librarian, I should grant my application some sort of up-to-date data basis to grant users. This led me to the following prompted conversation with my chosen LLM, ChatGPT 4.o:

## **USER**
I am designing an LLM application that could be capable of providing up to date information on house of representatives based on the user’s zip code, then providing further information on the users legislative activity, including bills, resolutions, roll calls, and committee memberships. Would you be able to provide this type of information in an up to date fashion with online search capabilities?

## **GPT 4.o's RESPONSE**

Yes, you can develop a large language model (LLM) application that provides up-to-date information on U.S. House of Representatives members based on a user's ZIP code, including their legislative activities such as bills, resolutions, roll call votes, and committee memberships. Here's how you can achieve this:

### 1. **Identify Representatives by ZIP Code**
To determine the congressional representative(s) for a given ZIP code:

* **Congress.gov**: Utilize the [Find Your Member](https://www.congress.gov/members/find-your-member) tool to input an address and retrieve the corresponding representative. This method is precise, especially when ZIP codes span multiple districts.

* **Geocodio API**: This API allows you to convert addresses or coordinates into congressional districts and append legislator contact information. It's particularly useful for bulk operations or integrating into applications .

* **WhoIsMyRepresentative.com API**: A straightforward API that returns data on representatives and senators based on ZIP code. It provides information such as name, state, district, phone, office, and website .

### 2. **Access Legislative Activity Data**
Once you've identified the representative, you can gather detailed legislative information:

* **Congress.gov API**: Offers comprehensive data on members, bills, resolutions, roll call votes, and committee memberships. You can retrieve information such as bill summaries, statuses, actions, and related committees .

* **Roll Call Votes**: Access detailed records of roll call votes to understand how representatives voted on specific legislation .

### 3. **Integrate Data into Your Application**
With the above APIs:

* **Data Retrieval**: Use the APIs to fetch real-time data based on user input (e.g., ZIP code).

* **Data Presentation**: Design your application to display the retrieved information in a user-friendly manner, allowing users to explore their representative's legislative activities.

* **Regular Updates**: Ensure your application periodically checks for updates to provide the most current information.

### 4. **Considerations**
* **ZIP Code Precision**: Be aware that some ZIP codes span multiple congressional districts. Using full addresses or integrating geolocation can improve accuracy.

* **API Rate Limits**: Monitor and adhere to the rate limits of the APIs you use to prevent service interruptions.

* **Data Privacy**: Ensure that any user data collected (like addresses) is handled securely and in compliance with privacy regulations.

By integrating these resources, your LLM application can effectively provide users with current and comprehensive information about their congressional representatives and their legislative activities.

## **Takeaways:** I found this conversation very helpful in starting my informed explorations of what this LLM application will be able to do, as seen in section 1. Overview and section 2. User Pathways. Some other APIs suggested to me, such as ProPublica API or whoismyrepresentative.com API, are either defunct or not publicly available. However, further research on Google Civic API showed promise in it's capability to find house representatives simply, something that Congress.gov cannot offer. Additionally, upon further research, Congress.gov is a more difficult format to work with, as it output XML format information, not JSON data formatting. It is also less publicly available and would require more acute knowledge of it's information, such as exact bill numbers opposed to bill names. Thus, I made the switch to GovTrack, a U.S. website that is open-source and tracks data on members of congress.

Now that I have my chosen LLM, user pathways, and a bridge to a database (aka, an API and open source data), where my trained LLM will extract its information to help users.
